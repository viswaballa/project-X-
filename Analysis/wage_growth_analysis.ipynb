```python
#import libraries
import pandas as pd
from pathlib import Path
import hvplot.pandas
```

```python
#set csv path and create wage dataframe
csvpath = Path('../Data/cleaned_wage_and_housing_data.csv')
wage_analysis_df = pd.read_csv(csvpath)
wage_analysis_df = wage_analysis_df.set_index('date')
```

```python
#display first 5 rows of dataframe
wage_analysis_df.tail()
```

```python
#group wage growth by year and find the cumulative % growth each year
wage_analysis_df.index=pd.to_datetime(wage_analysis_df.index)
cumulative_wage_growth_by_year = wage_analysis_df['wage_growth_yearly'].groupby(wage_analysis_df.index.year).sum()
print(cumulative_wage_growth_by_year)
```

```python
#group US housing price by year and find the cumulative % growth each year
housing_price_percent_change = wage_analysis_df['United States'].pct_change().dropna()*100
print(housing_price_percent_change)
cumulative_housing_price_growth_by_year = housing_price_percent_change.groupby(housing_price_percent_change.index.year).sum()
print(cumulative_housing_price_growth_by_year)
```

```python
#plot the cumulative change in housing prices and wage growth by year side by side
cumulative_housing_price_growth_by_year.hvplot.bar(label = "Cumulative monthly change in housing prices grouped by year (Feb 2016 - Jan 2023)",xlabel='Year',
ylabel='Cumulative monthly change') + cumulative_wage_growth_by_year.hvplot.bar(label = "Cumulative monthly change in wages grouped by year (Feb 2016 - Jan 2023)", xlabel='Year', ylabel='Cumulative monthly change') 
```

From the above, there appears to be somewhat of a directional trend between wages and housing prices. In years where the cumulative change in housing prices is above average, the change in wages is also generally above average, and vice versa. 

```python
#calculate the year over year change in housing prices
year_over_year_housing_prices = wage_analysis_df['United States'].pct_change(periods=11).dropna()*100
print(year_over_year_housing_prices)
```

```python
#combine the yearly changes in wages and housing prices into one df
combined_yearly_changes = pd.concat([wage_analysis_df['wage_growth_yearly'], year_over_year_housing_prices],axis='columns').dropna()
combined_yearly_changes = combined_yearly_changes.rename(columns={'United States':'housing prices', 'wage_growth_yearly':'wages'})
combined_yearly_changes.tail()
```

```python
#plot the year-over-year changes in housing prices and wages 
combined_yearly_changes.hvplot.line(label = "Year-over-year change (2017 - 2023)",xlabel='Month',
ylabel='y/y change (%)')
```

```python
#view the summary stats for the year-over-year changes
combined_yearly_changes.describe()
```

Housing prices have risen at a much faster clip than wages (on average, wages have increased 4% y/y each month over the past ~5 years versus 12% for housing prices). This suggests that housing affordability may be a increasing problem, as home prices have risen more than the income people have to buy them. The increase in housing prices has slowed but remains above that for wage growth, and there doesn't appear to be a strong statistical relationship between the two. 

```python
#see the correlation between y/y changes in wages and housing prices
combined_yearly_changes.corr()
```

There is a positive and very moderate correlation between y/y changes in wages and housing prices. However, we note it is not a long timeframe, with only 54 months in the sample
