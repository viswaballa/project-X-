```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
import hvplot.pandas
%matplotlib inline
```

```python
# Set csv path and create zillow housing price dataframe
sale_price = Path('Resources/median_sale_price_monthly.csv')
housing_df2 = pd.read_csv(sale_price, infer_datetime_format=True, parse_dates=True)
housing_df2
```

```python
# Remove columns except for RegionName and Dates 
housing_df2 = housing_df2.drop(columns=['RegionID', 'SizeRank', 'RegionType', 'StateName'])
housing_df2.sort_index(inplace=True)
housing_df2
```

```python
# Drops any date before 2018
housing_df2.drop(housing_df2.iloc[:, 1:118], inplace=True, axis=1)
housing_df2
```

```python
# Rename the Column here so that when I Transpose the column I have a
# easier time now than later trying to change 'RegionName'. 

housing_df2 = housing_df2.rename(columns={
    'RegionName': 'Date'
})
housing_df2
```

```python
# Use the `drop` function to drop specific column 'RegionName' because it was excessive 
housing_df2.set_index(housing_df2['Date'], inplace=True)
housing_df2
```

```python
# Dropped Date column since redundant 
housing_df2.drop(columns=['Date'], inplace=True)
housing_df2
```

```python
# Transpose the dataframe so that the TimeDate would be on the Indexed side. 
housing_df2 = housing_df2.T
housing_df2
```

```python
#Droping most of the cities that were not that big to make a smaller data set
housing_df2.drop(housing_df2.iloc[:, 60:535], inplace=True, axis=1)
```

```python
housing_df2.hvplot(
    rot=60,
    ylabel='Housing Index Price', 
    xlabel='Date',
    title='Housing Price Index from 2018 - 2022', 
).opts(yformatter='%.0f')
```
