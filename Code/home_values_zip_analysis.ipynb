```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
%matplotlib inline
```

```python
#Use the read_csv function and Path module to create a PD DataFrame with the imported uszips.csv file from Data folder
#And set zip as index

#uszips.csv was sourced from https://simplemaps.com/data/us-zips.

us_zip_codes = pd.read_csv(Path("../Data/uszips.csv", index_col="zip"))
us_zip_codes.set_index('zip', inplace=True)

home_values = pd.read_csv(Path("../Data/zillow_ZHVI_by_ZIP.csv", infer_datetime_format=True, parse_dates=True))

display(us_zip_codes.head())  #Review the first five rows of the zip codes DataFrame
display(us_zip_codes.tail())  #Review the last five rows of the zip codes DataFrame
                         
display(home_values.head())   #Review the first five rows of the home values DataFrame
display(home_values.tail())   #Review the last five rows of the home values DataFrame
```

```python
#Clean the uszips CSV and remove unnecessary columns such as timezone, military, county_name, county_weights, etc
us_zip_codes.drop(columns=['state_name', 'zcta', 'parent_zcta', 'density', 
                           'county_fips', 'county_name', 'county_weights', 
                           'county_names_all', 'county_fips_all', 'imprecise', 
                           'military', 'timezone'], inplace=True)
display(us_zip_codes.head())
display(us_zip_codes.tail())
```

```python
#Clean the zillow_ZHVI_by_ZIP CSV and remove unnecessary columns
home_values.drop(columns=['RegionID', 'SizeRank', 'RegionType',
                          'StateName', 'CountyName', 'Metro'], inplace=True)

display(home_values.head())
display(home_values.tail())
```

```python
start_idx = home_values.columns.get_loc("2017-12-31")  #Retrieve column index for 5-year start date of 1-1-2018
print("Start Index: "+ str(start_idx))
```

```python
home_values.drop(home_values.iloc[:, 3:219], inplace=True, axis=1)  #Drop dates before 2018
display(home_values.head())
```

```python
#Rename RegionName to Zip Code
home_values.rename(columns={"RegionName": "Zip Code"}, inplace=True) 
display(home_values.head())
```

```python
home_values_clean = home_values.dropna()

home_values_clean = pd.melt(home_values_clean, id_vars=['Zip Code', 'State', 'City'], value_vars=[cols for cols in home_values_clean.columns if cols not in ['Zip Code', 'State', 'City']])
home_values_clean
```

```python
#Rename columns
us_zip_codes.rename(columns={"index": "Zip Code", 
                             "lat": "Latitude", 
                             "lng": "Longitude",
                             "city": "City", 
                             "state_id": "State", 
                             "population": "Population"}, inplace=True) 
us_zip_codes.reset_index(inplace=True)  #Reset index

#Print sample DataFrame
display(us_zip_codes.head())
display(us_zip_codes.tail())
```

```python
#Merge zip code coordinate DataFrame with cleaned Home Value Index DataFrame
combined_df = us_zip_codes.merge(home_values_clean, left_on='zip', right_on='Zip Code', how="inner")

combined_df.dropna(inplace=True)  #Drop nulls
combined_df.isnull().sum()  #Validate nulls are dropped

display(combined_df.head())  #Print combined dataframe sample
display(combined_df.tail())  #Print combined dataframe sample
```

```python
#Drop unnecessary columns
combined_df.drop(columns=['Zip Code', 'State_y', 'City_y', 'index'], inplace=True)

#Create new DataFrame without population data since it only has values for one year rather than 5 years
zhvi_by_zip = combined_df.drop(columns=['Population'])

display(zhvi_by_zip.head())
display(zhvi_by_zip.tail())

```

```python
#Rename columns
zhvi_by_zip.rename(columns={"zip": "Zip Code",
                            "City_x": "City",
                            "State_x": "State",
                            "variable": "Date",
                            "value": "Home Value Index"}, inplace=True)

zhvi_by_zip.set_index('Zip Code')
zhvi_by_zip['Date'] = pd.to_datetime(zhvi_by_zip['Date'], errors='coerce')
grouped_zhvi = zhvi_by_zip.groupby([zhvi_by_zip.Date.dt.year, 'Zip Code'])['Home Value Index'].mean()
#zhvi_by_zip.set_index(zhvi_by_zip['Date'], inplace=True)
#zhvi_by_zip.drop(columns=['Date'], inplace=True)

#values_by_year_by_zip = zhvi_by_zip.groupby(['Year','Neighborhood']).mean()

print(grouped_zhvi.dtypes)
grouped_zhvi
#Print sample DataFrame
#display(grouped_zhvi.head())
#display(grouped_zhvi.tail())
```

```python
#Create DataFrames that slice the zip code data by city name
nyc_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'New York City') & (zhvi_by_zip['State'] == 'NY')]
la_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'Los Angeles') & (zhvi_by_zip['State'] == 'CA')]
chicago_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'Chicago') & (zhvi_by_zip['State'] == 'IL')]
houston_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'Houston') & (zhvi_by_zip['State'] == 'TX')]
phoenix_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'Phoenix') & (zhvi_by_zip['State'] == 'AZ')]
philly_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'Philadelphia') & (zhvi_by_zip['State'] == 'PA')]
san_antonio_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'San Antonio') & (zhvi_by_zip['State'] == 'TX')]
san_diego_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'San Diego') & (zhvi_by_zip['State'] == 'CA')]
dallas_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'Dallas') & (zhvi_by_zip['State'] == 'TX')]
san_jose_zip = zhvi_by_zip.loc[(zhvi_by_zip['City'] == 'San Jose') & (zhvi_by_zip['State'] == 'CA')]




display(nyc_zip)
display(la_zip)
display(chicago_zip)
display(houston_zip)
display(phoenix_zip)
display(philly_zip)
display(san_antonio_zip)
display(san_diego_zip)
display(dallas_zip)
display(san_jose_zip)
```
