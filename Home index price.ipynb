```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path
import hvplot.pandas
%matplotlib inline
```

```python
# Set csv path and create zillow housing price dataframe
sale_price = Path('Resources/median_sale_price_monthly.csv')
housing_df2 = pd.read_csv(sale_price, infer_datetime_format=True, parse_dates=True)
housing_df2
```

```python
# Remove columns except for StateName and Dates 
housing_df2 = housing_df2.drop(columns=['RegionID', 'SizeRank', 'RegionType', 'RegionName'])
housing_df2.set_index(housing_df2['StateName'], inplace=True)
housing_df2.sort_index(inplace=True)
housing_df2
```

```python
# Use the `drop` function to drop specific column 'StateName' because it was excessive 
housing_df2.drop(columns=['StateName'], inplace=True)
housing_df2
```

```python
# Transpose the dataframe so that the TimeDate would be on the Indexed side. 
housing_df2 = housing_df2.T
housing_df2.head()
```
