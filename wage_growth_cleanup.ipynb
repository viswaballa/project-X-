```python
#import libraries
import pandas as pd
from pathlib import Path
```

```python
#set csv path and create wage dataframe
csvpath = Path('project-X-/Data/fed_wage_growth_tracker.csv')
wage_df = pd.read_csv(csvpath, header = None)
```

```python
#display first 5 rows of dataframe
wage_df.head()
```

```python

#set the approrpriate row as the dataframe header
wage_df = wage_df.rename(columns=wage_df.iloc[1])
wage_df.head()
```

```python
#drop the top (uneeded) row of the dataframe
wage_df = wage_df.drop(0)
wage_df.head()
```

```python
#drop the other uneeded row of the dataframe (this row is now the header so it's not needed)
wage_df = wage_df.drop(1)
wage_df.head()
```

```python
#remove all columns except the 'unweighted overall' and 'date' columns
wage_df = wage_df[['Unweighted Overall', 'date']]
wage_df.head()
```

```python
#rename the column to wage_growth_yearly
wage_df = wage_df.rename(columns={'Unweighted Overall' : 'wage_growth_yearly'})
wage_df.head()
```

```python
#drop na's and check if there are any remaining
wage_df = wage_df.dropna()
wage_df.isna().mean()*100
```

```python
#infer datetime format and parse dates
wage_df['date'] = pd.to_datetime(wage_df['date'], infer_datetime_format=True)

#apply month end dates
wage_df['date'] = wage_df['date'].apply(lambda x:x +pd.offsets.MonthEnd(0))

#set the date as the index
wage_df = wage_df.set_index('date')
wage_df.head()
```

```python
#only keep rows with data from 2000 and later
cutoff_date = '1999-12-02'
wage_df = wage_df.loc[cutoff_date:]
wage_df.head()
```

```python
#set csv path and create zillow housing price dataframe
csvpath = Path('project-X-/Data/zillow_housing_price_index.csv')
housing_price_df = pd.read_csv(csvpath)
housing_price_df.head()
```

```python
#remove columns except the dates and region name
housing_price_df = housing_price_df.drop(columns=['RegionID', 'SizeRank', 'RegionType', 'StateName'])
housing_price_df.head()
```

```python
#transpose the dataframe
housing_price_df = housing_price_df.T
housing_price_df.head()
```

```python
#reset index
housing_price_df = housing_price_df.reset_index()
housing_price_df.head()
```

```python
#set first row as the header
housing_price_df = housing_price_df.rename(columns=housing_price_df.iloc[0])
housing_price_df.head()
```

```python
#remove the duplicate row
housing_price_df = housing_price_df.drop(0, axis=0)
housing_price_df.head()
```

```python
#rename the first column to 'date'
housing_price_df = housing_price_df.rename(columns={housing_price_df.columns[0]: 'date'})
housing_price_df.head()
```

```python
#drop any na's and check if there are any remaining
housing_price_df = housing_price_df.dropna()
housing_price_df.isna().mean()*100
```

```python
#infer datetime format
housing_price_df['date'] = pd.to_datetime(housing_price_df['date'], infer_datetime_format=True)

#apply month end dates
housing_price_df['date'] = housing_price_df['date'].apply(lambda x:x +pd.offsets.MonthEnd(0))

#set the date as the index
housing_price_df = housing_price_df.set_index('date')
housing_price_df.head()
```

```python
wage_and_housingprice_df = pd.concat([wage_df,housing_price_df], axis='columns', join='inner')
wage_and_housingprice_df.tail()
```

```python
#drop nas
wage_and_housingprice_df = wage_and_housingprice_df.dropna()
#check for nas
wage_and_housingprice_df.isna().mean()*100
```
